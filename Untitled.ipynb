```python
import math
from random import random
from random import seed
import numpy as np
```

```python
lose=0
win=0

tests=1000
np.random.seed(150)
good=np.random.rand(tests)
good=[math.ceil(a*3) for a in good]
selection=np.random.rand(tests)
selection=[math.ceil(a*3) for a in selection]
for a in range(tests):
    ns=0
    for s in [1.0, 2.0, 3.0]:
        if s!=good[a] and s!=selection[a]:
            ns=s
    print "You selected door #"+str(int(selection[a]))+" and door #"+str(int(ns))+" has carrots! Will you change?"
    if good[a]==selection[a]:
        print "If you change you lose! the prize was behind door #"+str(int(good[a]))
        lose=lose+1
    if good[a]!=selection[a]:
        print "If you change you win! the prize was behind door #"+str(int(good[a]))  
        win=win+1
    print ""
print "Changers won "+str(win)+"/"+str(tests)+" v. non-changers "+str(tests-win)+"/"+str(tests)
```

```python
def fibo(counter, a=1, b=1):
    if counter>0:
        return fibo(counter-1, b, a+b)
    else:
        return a

r=fibo(9)
print r
```

```python
def fib(n):
    if n==0:
        return 1
    if n==1:
        return 1
    return fib(n-1)+fib(n-2)
r=fib(9)
print r
```
